## Function Definitions

```python
from scikits.audiolab import Sndfile
import numpy as np
import pprint
pp = pprint.PrettyPrinter(indent=4)
from numpy.fft import fft,fftfreq

def audio2Farray(audio_file):
    ''' Takes path to audiofile and returns an ndarray containing the positive
    portion of sample's fourier transform, with a corresponding array of 
    frequencies. If the file has multiple channels, only the first channel is used.'''
    f = Sndfile(audio_file, 'r')

    # Sndfile instances can be queried for the audio file meta-data
    fs = f.samplerate
    nc = f.channels
    enc = f.encoding
    nf = f.nframes

    # Default, read in as float with values normalized from 0 to 1
    data = f.read_frames(nf)
    #data.shape
    
    # take data from the 0th channel
    data1 = data[:,0] #.astype(np.int32)
    
    Fdata_full = fft(data1)
    Fdata_full.shape
    freq_full = fftfreq(Fdata_full.shape[0], d=time.max()/float(data1.shape[0]) )

    # remove data at negative frequencies, and take the absolute value
    Fdata = np.abs(Fdata_full[:np.ceil(Fdata_full.shape[0]/2)])
    freq = np.abs(freq_full[:np.ceil(Fdata_full.shape[0]/2)])
    
    return Fdata, freq
```

```python
def findPeaks(F,freq,amp_tol):
    ''' Algorithm to find peaks by searching frequeny ranges for a maximum above some 
    fraction of the global maximum amplitude. Returns the elements in notes for which this 
    criteria holds. Returns list of tuples containing (fit frequency, frequency of closest
    note, label of closest note)'''
    global notes
    
    assert(amp_tol > 0. and amp_tol <= 1.)
    max_amp = F.max()
    tol = amp_tol*max_amp
    
    peaks = []
    for note in notes:
        frange = note[2]
        idx =  np.logical_and(freq>=frange[0], freq<=frange[1])
        nearnote = F[idx]
        freq_nearnote = freq[idx]
        if nearnote.max() > tol:
            freq_peak = freq_nearnote[nearnote.argmax()]
            # save as (fit freq, target freq, note label, note range, max amplitude)
            peaks.append((freq_peak,note[0],note[1],note[2],nearnote.max()))
            
    return peaks

```

```python
from copy import copy

noteOrder = [ s[1][0] for s in scale]
def detectBleed(peaklist,width=0.25):
    '''Detect situation where the neighboring note is falsely detected, by checking that
    the neighboring note has a lower maximum, and that maxium is in the nearest portion
    of size 'width' in its range to the correct note.'''
    global noteOrder
    
    assert(0.<=width<=1.)
    
    peaks = copy(peaklist) # make a copy
    removelist = []
    for i in range(len(peaks)-1):
        a = peaks[i]; b = peaks[i+1]
        
        # determine whether two consecutive peaks are neighboring notes
        # or identically whether the they share a boundary on their search range
        if  a[3][1] == b[3][0]:
            
            # check for a bleeding into b
            if a[4] > b[4] and abs(b[0] - b[3][0]) < abs( b[3][1] - b[3][0])*width:
                removelist.append(i+1)
            # check for b bleeding into a
            elif a[4] < b[4] and abs(a[0] - a[3][1]) < abs( a[3][1] - a[3][0])*width:
                removelist.append(i)
                
        # reverse list before pop()
        removelist.reverse()
        
        # Remove false positives from peaks and save in separate list
        removed = []
        for idx in removelist:
            removed.append(peaks.pop(idx))
            #print 'Detected false-positive: {}{}'.format(removed[-1][2][0],removed[-1][2][1])
            
        removed.sort()
        return peaks, removed
                    
```

```python
from copy import copy
def removeHarmonics(peakslist,width=.5):
    ''' Remove any higher harmonics from a list of identified peaks. The
    tolerance for finding a harmonic is:
        ('width') x ( search range of fundamental ) x ( harmonic order )
        
    Note: This method will remove correct notes that happen to be a harmonic
    of a lower note (such as playing notes an octave apart'''
    assert(0.<=width<=1.)

    peaks = copy(peakslist) # make copy of list
    fundamentals = []
    harmonics = []
    notes_detected = []
    while len(peaks) > 0:
        fund = peaks.pop(0)
        maxmult = np.ceil(5000. /fund[1])
        harms = fund[1]*np.arange(2,maxmult)
        
        harm_idx = []; harm_order = [];
        for i,peak in enumerate(peaks):
            for j,harm in enumerate(harms.tolist()):
                tol = abs(fund[3][0]-fund[3][1])*width*j
                #print fund[1]*j, peak[0], tol
                # check for jth order harmonic
                if abs(peak[0] - fund[1]*j) < tol:
                    harm_idx.append(i)
                    harm_order.append(j)
                    #print '{}th Harmonic of {}{} identified'.format(j,fund[2][0],fund[2][1])
                    #print fund[1]*j, peak[0], tol

        fundamentals.append(fund)
        
        # reverse lists for pop()
        harm_idx.reverse()
        harm_order.reverse()
        for idx,order in zip( harm_idx,harm_order):
            harm = peaks.pop(idx)
            # save as ( fit freq, inferred freq, corresponding fundament, order, alternative)
            harmonics.append( (harm[0],fund[1]*order,fund[2],order, harm[2]) )
        harmonics.sort()        
                
    return fundamentals,harmonics
                    
```

```python
#testfile = './sound_files/F3_PopOrgan.aif'
#testfile = './sound_files/A4_PopOrgan.aif'
testfile = './sound_files/C4+A4_PopOrgan.aif'
#testfile = './sound_files/F4_CathedralOrgan.aif'

Fdata, freq = audio2Farray(testfile)
mypeaks = findPeaks(Fdata,freq,0.1)
print 'All detected peaks:'
print(mypeaks)

truePeaks,falsePositives = detectBleed(mypeaks)
myfund,myharm = removeHarmonics(truePeaks)

print('\nResults:')

print('\n\tUnique Fundamental Notes:');print '\t',
for fund in myfund: print '{:.1f}-{}{}'.format(fund[0],fund[2][0],fund[2][1]),

print('\n\tRuled out as false-positives:');print '\t',
for fp in falsePositives: print '{:.1f}-{}{}'.format(fp[0],fp[2][0],fp[2][1]),

print('\n\tRuled out as harmonics:');print '\t',
for fp in myharm: print '{:.1f}-{}{}({}*{}{})'.format(fp[0],fp[4][0],fp[4][1], fp[3],fp[2][0],fp[2][1]),
```
